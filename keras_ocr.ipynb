# We try first reading a license plate taken closely

```
zoomed_plate = Image.open( 'zoomed plate.JPG' )
zoomed_plate
```

```
image1 = keras_ocr.tools.read('zoomed plate.JPG')
pipeline = keras_ocr.pipeline.Pipeline()
prediction_image1 = pipeline.recognize([image1])
```

```
keras_ocr.tools.drawAnnotations(image=image1, predictions=prediction_image1[0])
```

## As we can see, the recognition phase succeeded

# Now, we are going to read a license plate picture taken from far away  

```
plate = Image.open( 'plate.JPG' )
plate
```

```
image2 = keras_ocr.tools.read('plate.JPG')
pipeline = keras_ocr.pipeline.Pipeline()
prediction_image2 = pipeline.recognize([image2])
```
